## 2. Heterogeneity Analysis with local L matrix
### i. Calculate the local L matrix
For heterogeneity analysis, we will need the local L matrix.  
The **local L matrix** serves as the single-cell/pseudo-bulk regulatory state indicator.  
Like the transcriptome matrix, we can do all kinds of analyses with the **regulatory state matrix**.  


By calling ``graph_measure.Local_L( )``, the local L matrix will be calculated with user-specified arguments.  
  
- **I_cutoff** sets the cutoff for Moran's I. If gene or peak's Moran's I is below the cutoff, pair will be filtered out. 
- **pct_cutoff** sets the feature sparsity cutoff. Default is 0.05, meaning that features should be > 0 in at least 5% of the cells/pseudo-bulks.
- **dropout_rm** specifies whether to make local L index to be 0 if feature value is 0 (dropout). Default is True (recommended)


```python
toy = graph_measure.Local_L(toy,
                            I_cutoff=0.2,
                            pct_cutoff=0.05,
                            dropout_rm=True)
```

### ii. Sub-clustering and trajectory inferring
Next, we will do subclustering and trajectory inferring with the regulatory state matrix by calling ``RegulatoryAnalysis( )`` with user specified arguments.  
  
- **binary**: whether to use the binarized local L matrix or not. Default is False.

- **weighted**: whether assign weights by feature variance to the local L matrix or not. Default is False.

- **trajectory**: if True, do the trajectory analysis after sub-clustering

- **leiden_resol**: resolution in sub-clustering

- **root**: cluster to use as the root in trajectory.


```python
import scgreat.analysis as scanalysis
```

In the very first run, we will not specify the root cluster since we haven't got the cluster yet.

```python
toy = scanalysis.RegulatoryAnalysis(toy, binary=False,
                                    weighted=False, leiden_resol=0.15)
```

With the trajectory shown above, we can now choose a root cluster. For this toy example, let's just choose 0.

```python
toy = scanalysis.RegulatoryAnalysis(toy, binary=False, root='0',
                                    weighted=False, leiden_resol=0.15)
```

Finally, we can visualize the results with Scanpy.

```python
sc.pl.umap(toy, size=50, color=['leiden','dpt_pseudotime'])
```

### iii. Feature module discovery

With the trajectory inferred, we can now cluster all the features based on their temporal patterns by calling ``Time_Module( )`` with user specified arguments.  
  
- ``features``: feature list to cluster
  
  
- ``pseudotime_label``: column name in anndat.obs containing pseudo-time labels 
  
  
- ``bins``: split the pseudo-timeline into N bins, use the average value within each bin for module discovery
  
  
- ``num_iteration``: maximum number of iteration to optimize the Self-Oranizing Map (SOM)
  
  
- ``som_shape``: (M, N) shape of the map, defines number and similarity structure of modules
  
  
- ``sigma``: the radius of the different neighbors in the SOM
  
  
- ``learning_rate``: optimization speed, how much weights are adjusted during each iteration
  
  
- ``module_name``: SOM name by the user, if not provided, 1, 2, 3, 4, etc., will be used
  
  
- ``random_seed``: random seed to make the results reproducible
  

Here we will cluster the peaks in the toy example, genes can also be applied to the discovery function.

```python
features = [gp.split('_')[1] for gp in toy.uns['Local_L_names']]
toy = scanalysis.Time_Module(toy, features,
                                          bins=25,
                                          pseudotime_label='dpt_pseudotime',
                                          som_shape=(1,2))
```

```python
import scgreat.visualization as visualization
```

```python
f = visualization.visualize_module(toy, '1', figsize=(8,4))
```

### iv. Cluster marker discovery

When cluster labels are avialable, whether it is from the user or previous analysis, we can now discover differentially correlated gene-peak pairs among clusters. It is similar as Differentially Expressed Genes (DEGs) analysis, but the input is our local L matrix.  
  
Marker discovery can be done with ``FindAllMarkers`` for all groups markers or ``FindMarkers`` for two-group comparison (e.g., disease vs normal). 

- ``ident``: column name in AnnData Object containing group labels  
  
  
- ``corrct_method``: multi-test correction method, one of ['bonferroni', 'fdr']  
  
  
- ``seed``: random seed to make the results reproducible
  
  
- ``group_1`` (``FindMarkers`` only): first group name in ident to compare with the second
  
  
- ``group_2`` (``FindMarkers`` only): second group name in ident to compare with the first
  
  
\* Before calling the function below, it's recommended to add the group-specific feature sparsity of original single-cell data to the multiome object for quality control. This is uaually done in the data preparation step by calling ``add_group_sparsity( )``

```python
toy, stat_df_all = scanalysis.FindAllMarkers(toy,
                                             ident = 'leiden',
                                             corrct_method='bonferroni',
                                             seed=1) #11075
```

When the function was called, 
- it will first calculate per group local L matrices for the comparison (you can also run ``Local_L_bygroup`` before calling the function to tune the cutoffs)
- then perform statistical test
- finally return the test results as DataFrame ``stat_df_all``

```python
stat_df_all
```

To select significant pairs, call ``MarkerFilter`` with user-specified arguments.  
  
- ``stat_df``: differentially regulated pairs statistical test results
  
  
- ``min_pct_rna``: percentage of cells that express the gene as sparsity cutoff
  
  
- ``min_pct_atac``: percentage of cells that have the peak as sparsity cutoff
  
  
- ``mean_diff``: mean correlation difference between the group and background (all other groups)
  
  
- ``p_cutoff``: adjusted p-value cutoff
  
  
- ``plot``: if True, return volcano plot
  
  

```python
stat_df = scanalysis.MarkerFilter(stat_df_all,
                                  min_pct_rna=0,
                                  min_pct_atac=0,
                                  mean_diff=0.15,
                                  p_cutoff=1e-2,
                                  plot=True)
```

### v. Motif enrichment analysis

With peaks of interested from any above analyses, the package helps the users to call Homer from inside the Python process:  
- It will prepares input file with Homer required format
- It can help extract the peak information when motif of interests is identified from Homer results

Here, we will take the markers we found just now for an example.  
To run Homer, call ``run_HOMER_motif`` with arguments below:  

- ``peaks``: list or array-like, peaks of interests
  
  
- ``out_dir``: output directory to save the results
  
  
- ``prefix``: prefix of all files, folder called ``out_dir/homer_prefix`` will be created to save all the results
  
  
- ``ref_genome``: str, reference genome name, e.g., 'hg19', 'hg38'
  
  
- ``homer_path``: path to Homer software, if Homer already added to the PATH, argument can be ignored
  
  
- ``split_symbol``: how peak location ID is merged
                      'chr1-12345-23456' - split_symbol=['-','-']
                      'chr1:12345-23456' - split_symbol=[':','-']
  
  
- ``size``: Homer paramter, the size of the region used for motif finding
  
  
\* Please make sure you have Homer installed and know the path of it. Tutorials can be found here: http://homer.ucsd.edu/homer/introduction/install.html

```python
peaks = [f.split('_')[1] for f in stat_df['name']]
peaks = np.unique(peaks)
print(len(peaks))
homer_df = scanalysis.run_HOMER_motif(peaks, out_dir='homer/',
                                      prefix='PREFIX', ref_genome='hg38',
                                      homer_path='PATH/TO/HOMER',
                                      split_symbol=['-','-'], size=200)
```
