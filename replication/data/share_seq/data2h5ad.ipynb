```python
import numpy as np
import pandas as pd
import scipy.io
import scanpy as sc
import anndata as ad
```

### scRNA-seq

```python
rna_counts = pd.read_table("raw/GSM4156608_skin.late.anagen.rna.counts.txt.gz", index_col=0)
rna_obs = pd.DataFrame(index=rna_counts.columns)
rna_obs.index = rna_obs.index.str.replace(",", ".")
rna_var = pd.DataFrame(index=rna_counts.index)
rna_obs.index.name, rna_var.index.name = "cells", "genes"
rna = ad.AnnData(
    X=scipy.sparse.csr_matrix(rna_counts.to_numpy().T),
    obs=rna_obs,
    var=rna_var
)
rna
```

```python
rna.obs["domain"] = "scRNA-seq"
rna.obs["protocol"] = "SHARE-seq"
rna.obs["dataset"] = "Ma-2020-RNA"
rna.var["genome"] = "mm10"
```

### scATAC-seq

```python
atac_counts = scipy.io.mmread("raw/GSM4156597_skin.late.anagen.counts.txt.gz")
atac_obs = pd.read_table(
    "raw/GSM4156597_skin.late.anagen.barcodes.txt.gz",
    header=None, names=["Cells"], index_col=0
)
atac_var = pd.read_table(
    "raw/GSM4156597_skin.late.anagen.peaks.bed.gz",
    header=None, names=["chrom", "chromStart", "chromEnd"]
)
atac_obs.index.name, atac_var.index.name = "cells", "peaks"
atac = ad.AnnData(
    X=atac_counts.T.tocsr(),
    obs=atac_obs,
    var=atac_var
)
atac
```

```python
atac.obs["domain"] = "scATAC-seq"
atac.obs["protocol"] = "SHARE-seq"
atac.obs["dataset"] = "Ma-2020-ATAC"
```

```python
atac.var.index = pd.Index(
    atac.var["chrom"] + ":" +
    atac.var["chromStart"].astype(str) + "-" +
    atac.var["chromEnd"].astype(str),
    name=atac.var.index.name
)
atac.var["genome"] = "mm10"
```

```python
atac.X.max()
```

### Clean data

```python
rna = rna[atac.obs_names]
np.all(rna.obs_names == atac.obs_names)
```

```python
sc.pp.filter_genes(rna, min_counts=10)
rna
```

```python
sc.pp.filter_genes(atac, min_counts=10)
atac
```

```python
rna.X = rna.X.toarray()
```

```python
rna.write_h5ad('SHARE_mouseskin_RNA.h5ad')
atac.write_h5ad('SHARE_mouseskin_ATAC.h5ad')
```

### Preprocess

```python
anndat_mod1 = ad.read_h5ad('SHARE_mouseskin_RNA.h5ad')
anndat_mod2 = ad.read_h5ad('SHARE_mouseskin_ATAC.h5ad')

sc.pp.normalize_total(anndat_mod1)
sc.pp.log1p(anndat_mod1)
print(anndat_mod1.X.max())
sc.pp.highly_variable_genes(anndat_mod1)
anndat_mod1.raw = anndat_mod1

anndat_mod1 = anndat_mod1[:, anndat_mod1.var.highly_variable]
sc.pp.scale(anndat_mod1, max_value=10)
sc.tl.pca(anndat_mod1, svd_solver='arpack', use_highly_variable=False)
sc.pp.neighbors(anndat_mod1, n_neighbors=25, n_pcs=30)
sc.tl.leiden(anndat_mod1, resolution=0.5)
sc.tl.umap(anndat_mod1)
```

```python
sc.set_figure_params(
    scanpy=True, dpi_save=600, vector_friendly=True, format="pdf",
    facecolor=(1.0, 1.0, 1.0, 0.0), transparent=False
)

sc.pl.umap(anndat_mod1,color='leiden')
```

```python
anndat_mod1_raw = ad.read_h5ad('SHARE_mouseskin_RNA.h5ad')
anndat_mod1_raw.obs = anndat_mod1.obs
anndat_mod1_raw.uns = anndat_mod1.uns
anndat_mod1_raw.obsm = anndat_mod1.obsm
anndat_mod1_raw.obsp = anndat_mod1.obsp

anndat_mod2.obs = anndat_mod1_raw.obs
```

```python
anndat_mod1_raw.write_h5ad('SHARE_mouseskin_RNA.h5ad')
anndat_mod2.write_h5ad('SHARE_mouseskin_ATAC.h5ad')
```
