```python
import numpy as np
import pandas as pd
import networkx as nx
import scipy.io
import scanpy as sc
import anndata as ad
```

### scRNA-seq

```python
rna_matrix = scipy.io.mmread("raw/GSE126074_AdBrainCortex_SNAREseq_cDNA.counts.mtx.gz").T.tocsr()
rna_obs = pd.read_csv("raw/GSE126074_AdBrainCortex_SNAREseq_cDNA.barcodes.tsv.gz", header=None, index_col=0)
rna_var = pd.read_csv("raw/GSE126074_AdBrainCortex_SNAREseq_cDNA.genes.tsv.gz", header=None, index_col=0)
rna_obs.index.name, rna_var.index.name = "cells", "genes"
rna = ad.AnnData(X=rna_matrix, obs=rna_obs, var=rna_var)
rna
```

```python
rna.obs["domain"] = "scRNA-seq"
rna.obs["protocol"] = "SNARE-seq"
rna.obs["dataset"] = "Chen-2019-RNA"
```

```python
rna.X = rna.X.toarray()
```

```python
rna.write_h5ad('SNARE_mousebrain_RNA.h5ad')
```

### scATAC-seq

```python
atac_matrix = scipy.io.mmread("raw/GSE126074_AdBrainCortex_SNAREseq_chromatin.counts.mtx.gz").T.tocsr()
atac_obs = pd.read_csv("raw/GSE126074_AdBrainCortex_SNAREseq_chromatin.barcodes.tsv.gz", header=None, index_col=0)
atac_var = pd.read_csv("raw/GSE126074_AdBrainCortex_SNAREseq_chromatin.peaks.tsv.gz", header=None, index_col=0)
atac_obs.index.name, atac_var.index.name = "cells", "peaks"
atac = ad.AnnData(X=atac_matrix, obs=atac_obs, var=atac_var)
atac
```

```python
atac.obs["domain"] = "scATAC-seq"
atac.obs["protocol"] = "SNARE-seq"
atac.obs["dataset"] = "Chen-2019-ATAC"
```

```python
atac.X.max()
```

```python
atac.write_h5ad('SNARE_mousebrain_ATAC.h5ad')
```

### Preprocess

```python
anndat_mod1 = ad.read_h5ad('SNARE_mousebrain_RNA.h5ad')
anndat_mod2 = ad.read_h5ad('SNARE_mousebrain_ATAC.h5ad')

sc.pp.normalize_total(anndat_mod1)
sc.pp.log1p(anndat_mod1)

sc.pp.highly_variable_genes(anndat_mod1)
anndat_mod1.raw = anndat_mod1

anndat_mod1 = anndat_mod1[:, anndat_mod1.var.highly_variable]
sc.pp.scale(anndat_mod1, max_value=10)
sc.tl.pca(anndat_mod1, svd_solver='arpack', use_highly_variable=False)
sc.pp.neighbors(anndat_mod1, n_neighbors=25, n_pcs=30)
sc.tl.leiden(anndat_mod1, resolution=0.5)
sc.tl.umap(anndat_mod1)
```

```python
sc.pl.umap(anndat_mod1,color='leiden')
```

```python
anndat_mod1_raw = ad.read_h5ad('SNARE_mousebrain_RNA.h5ad')
anndat_mod1_raw.obs = anndat_mod1.obs
anndat_mod1_raw.uns = anndat_mod1.uns
anndat_mod1_raw.obsm = anndat_mod1.obsm
anndat_mod1_raw.obsp = anndat_mod1.obsp

anndat_mod2.obs = anndat_mod1_raw.obs
```

```python
anndat_mod1_raw.write_h5ad('SNARE_mousebrain_RNA.h5ad')
anndat_mod2.write_h5ad('SNARE_mousebrain_ATAC.h5ad')
```
